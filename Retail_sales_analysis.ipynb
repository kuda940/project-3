```python
def age_to_group(age):
    if age < 20:
        return 'Teen'
    elif age < 30:
        return '20s'
    elif age < 40:
        return '30s'
    else:
        return '40+'

df['AgeCategory'] = df['Age'].apply(age_to_group)
category_groups = df.groupby('AgeCategory')
```

# Task
Analyze the provided sales data to identify seasonal trends and patterns.

## Extract time components

### Subtask:
Extract the year and month from the 'Date' column and create new columns for 'Year' and 'Month'.


**Reasoning**:
Extract the year and month from the 'Date' column and create new columns for 'Year' and 'Month'.



```python
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
display(df.head())
```

## Group and aggregate data

### Subtask:
Group the DataFrame by 'Year' and 'Month' and calculate the sum of 'Total Amount' for each group.


**Reasoning**:
Group the DataFrame by 'Year' and 'Month' and calculate the sum of 'Total Amount' for each group, then reset the index and store the result in `monthly_sales`.



```python
df['day_of_week'] = df['Date'].dt.day_name()
display(df.head())
```

```python
#visualize by quater
```

```python
df['Quarter'] = df['Date'].dt.quarter
quarterly_sales = df.groupby(['Year', 'Quarter'])['Total Amount'].sum().reset_index()
display(quarterly_sales)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=quarterly_sales['Year'].astype(str) + '-Q' + quarterly_sales['Quarter'].astype(str), y='Total Amount', data=quarterly_sales)
plt.title('Quarterly Sales')
plt.xlabel('Quarter')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
monthly_sales = df.groupby(['Year', 'Month'])['Total Amount'].sum().reset_index()
display(monthly_sales)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str), y='Total Amount', data=monthly_sales)
plt.title('Monthly Sales Trends')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
